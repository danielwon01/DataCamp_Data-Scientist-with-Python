# Regrssion

#### 회귀 분석은 둘 이상의 변수 간의 관계를 보여주는 통계적 방법입니다. 
 머신러닝에서의 회귀는 여러 개의 독립변수와 한 개의 종속변수 간의 상관관계를 모델링 하는 것이 기본적인 개념이며 독립변수에 영향을 미치는 회귀 계수(Regression coefficients)의 최적 값을 찾아 종속변수를 예측



##### 단순선형회귀란 독립변수(X) 와 종속변수 (Y)가 하나 씩 존재하는 회귀를 의미

Y = 예측값    
W = 모델계수 / feature wights                      
b = 편중항 / 절편                                
e = 오류값

독립변수 - 입력값이나 원인을 나타낸다.  
종속변수  - 결과물이나 효과를 의미

#### 선형회귀 모델의 장단점 

###### 장점 
* 학습,예측 속도가 빠르다.
* 방대한 양의 데이터셋에도 학습이 잘 이루어진다.
* 특성이 많을 때 학습이 잘 될 수 있으며 예측이 어떻게 이루어지는지 쉽게 이해 가능 


###### 단점 
* 계수의 값들이 왜 그러한 값을 가지는지 명확하지 않다.
* 데이터셋의 특성이 서로 깊게 연관되어 있다면 계수를 분석하기 어려움 


##### 사용예 
 * 선형 모델은 샘플에 비해 특성이 많을 때 잘 작동한다. 

### 최적의 회귀 모델을 생성하는 방법 
##### 최적의 회귀 모델을 생성한다는 말의 의미는 잔차(residuals) 의 합(RSS, residual sum of squares)이 최소가 되는 선형식을 도출하는 것을 의미합니다. 설명한 최적의 회귀 계수를 찾는 것이라고 봐도 무방

* residuals   - 표본집단에서 회귀식을 얻었다면, 그 회귀식을 통해 얻은 예측값과 실제 관측값의 차이가 잔차
* error -  모집단에서 회귀식을 얻었다면, 그 회귀식을 통해 얻은 예측값과 실제 관측값의 차이


* 잔차는 +, - 값이 모두 가능하기 때문에 잔차를 단순히 더하게 되면 0이 되는 경우가 발생할 수 있어 일반적으로는 절댓값을 취하거나 
  MAE, Mean Abosoulte Error , 제곱합의 평균을 사용 RSS , Residual Sum of Square

#### 손실함수(비용함수 , Cost Fuction)  
#### 예측 값이 실제 값과 얼마나 유사한지 판단하는 기준 , 예측값과 실제 값의 차이를 loss라고 하며, 이를 줄이는 방향으로 학습이 진행 

#### 평균절대오차  ( MAE, Mean Abosoulte Error)
#### 예측값과 실제 값의 차이에 절대값을 취하여, 그 값들을 모두 더하고 개수로 나누어 평균을 낸 값 , n = 오차의 갯수 , |Xi - X|  = 절대 오차 
###### MAE는 에러의 절대값을 취하기 때문에 에러의 크기 그대로 반영 , 에러에 따른 손실이 선형적으로 올라갈때와 이상치가 많을때  사용 적합 
##### 절대값을 취하기 때문에 해당 예측이 어떤 식으로 오차가 발생했는지, 음수인지 양수인지 판단할 수 없는 단점이 존재


##### 평균제곱오차(MSE , Mean Squared Error)
##### 예측값과 실제값 사이의 평균을 제곱하여 평균을 낸 값,  차이가 커질 수록 제곱 연산으로 인해 값이 가시화 되며 제곱으로 인해 오차가 양수, 음수이든 누적값을 증가 시킨다.
**실제값에 대한 정답률의 오차뿐만 아니라 다른 오답들에 대한 정답률 또한 포함하여 계산한다는 것이 특징**

##### MAE와 달리 최적 값에 가까워 질 수록 이동거리가 다른게 변화하기 때문에 최적 값에  수렴하기 용이 하지만 값을 제곱하기 때문에 값 자체의 왜곡이 있을 수 있다는 단점 존재 


#### 평균제곱근오차(RMSE , Root Mean Squre Error)
##### MSE에 루트를 씌운 지표 ,  루트를 취하기 때문에 값을 제곱해서 생기는 왜곡이 줄어들며 , 오차를 보다 직관적으로 보여준다. 즉 오류값과 실제 값과 유사한 단위로 변환하여 해석할 수 있기 때문 



### 선형 모델 정규화(Regularization)
###### 정규화(regularized) 선형회귀 방법은 선형회귀 계수(weight)에 대한 제약 조건을 추가함으로써 모형이 과도하게 최적화되는 현상, 즉 과최적화를 막는 방법

##### L1 Norm(Mahattan Distance) 은 두개의 벡터를 빼고 , 절대값을 취한 뒤 합한것    
* x = (1,2,3) , y = (-4,-5,-6) -> d(x,y) = |1-(-4)|+|2-(-5)|+|3-(-6)| = 5+7+9

##### L2 Norm(Euclidean Distance) L2 Norm은 두 개의 벡터의 각 원소를 빼고, 제곱을 하고, 합치고, 루트를 씌운 것 
* x=(1,2,3), y=(-4,-5,-6)은 d(x,y)=root(5+7+9)

#### 능형회귀 (Ridge regression) -  L2 Regularization 
##### 최소제곱법과 매우 유사하나, 각 계수의 제곱을 더한 값을 식에 포함하여 계수의 크기도 함께 최소화하도록 만들었다는 차이 가 있다.
#####  분산이 큰 상황을 막기 위해, 계수 자체가 크면 패널티를 주는 수식을 추가한 것

#### Ridge 와 Lasso 차이점 
Ridge 회귀는 모델에서 중요한 기능이 많이있을 때 사용하는 것이 더 좋다. 모델에 페널티를 주지만 덜 중요한 기능은 삭제하지 않기 때문입니다. 올가미는 모델에서 가장 중요한 기능 만 사용하려고 할 때 더 잘 사용된다.

Lasso - 제약조건이 절대값이라 아래의 그림처럼 마름모꼴의 형태로 나타난다.**최적값은 모서리 부분에서 나타날 확률이 높아 유의미하지 않은 변수들에 대해 계수를 0 에 가깝게 추정해 주어 변수 선택 효과를 가져오게 됩니다.** 파라미터의 크기에 관계없이 같은 수준의 Regularization을 적용하기 때문에 작은 값의 파라미터를 0으로 만들어 해당 변수를 모델에서 삭제하고 따라서 모델을 단순하게 만들어주고 해석에 용이하게 만들어줍니다.

Ridge - **변수들 간의 공선성(회귀분석의 전제가정인 독립변수들 간에 상관관계가 높으면 안된다는 조건을 위배하는 경우를 의미)이 있을 경우 변수가 아무리 많아도 변수간 상관성 때문에 실제로 사용할 수 있는 정보가 적을 경우 이 독립변수들간의 분산을 줄이기 위해 사용**
