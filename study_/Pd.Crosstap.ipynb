#### pd.crosstap(index, columns, values=None, rownames=None, colnames=None, aggfunc=None, margins=False,                                                                 margins_name='All', dropna=True, normalize=False)   

범주형 변수로 되어있는 요인(factors)별로 교차분석(cross tabulations) 해서, 행, 열 요인 기준 별로 빈도를 세어서 도수분포표(frequency table), 교차표(contingency table) 를 만들어주는 pd.crosstab() 

```python
import pandas as pd
```

```python
df = pd.DataFrame({"x": [1,1,1,4,5], "y": [5,5,5,9,10], "z": [15,1,1,12,11]})
df
```

```python
# pd.crosstab()의 행과 열 위치에는 array 형식의 데이터가 들어간다.

pd.crosstab(df.x,df.y)
```

```python
pd.crosstab(df.y, df.z)
```

```python
# 행, 열 이름 변환
pd.crosstab(df.x ,df.z , rownames = ["A"] ,colnames = ["B"] )
```

```python
# 교차표의 행 합 , 열합 추가 
pd.crosstab(df.z, df.y , margins = True)
```

```python
# 구성비율로 교차표 만들기 
pd.crosstab(df.x , df.z , normalize = True)`
```

```python
# values ,aggfunc 계산 추가
pd.crosstab(df.x,df.z , values = df.z , aggfunc = "mean").round(0)
```
