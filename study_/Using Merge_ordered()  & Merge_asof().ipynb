# merge_ordered()

> 시계열 및 기타 정렬된 데이터를 병합 할 수 있다.  
  결과는 표준 병합방법과 유사하지만 결과가 정렬되서 나온다.
  
> merge_ordered() 들어가는 인수는 merge()와 동일하지만 merge는 inner join이 default 지만 merge_ordered은 outer join이 default

> pd.merge_odered(df,df1)

```python
import numpy as np
import pandas as pd
```

### 주식데이터 불러오기

```python
import pandas_datareader as pdr
apple = pdr.get_data_yahoo('aapl', start = '2021-10-01', end ='2021-10-31')
ms = pdr.get_data_yahoo('msft', start = '2021-10-01', end ='2021-10-31')

display(apple.head(3),  ms.head(3))
```

NULL 값 확인

### merge_ordered()

```python
ap_ms = pd.merge_ordered(apple,ms)
ap_ms.head()
```

```python
ap_ms_1 = pd.merge_ordered(apple,ms, on ='Date')  
ap_ms_1.head(3)
```

# merge_asof() 

> 동일한 값이 아닌 가장 가까운 값 열에서 병합 (병합되는 col들은 정렬되 있어야한다.) 

>결과값은 left join과 유사하게 왼쪽 테이블의 모든 행이 표시 ,오른쪽 테이블은 왼쪽테이블과 일치하는 정도를 기반으로 반환 

> pd.merge_asof(df,df1, direction = 'forward'') on에 지정한 키열이 왼쪽의 키 열보다 크거나 같은 오른쪽 테이블의 값 

> pd.merge_asof(df,df1, direction = 'backword') 키열과 같거나 작은 값 (default)

> pd.merge_asof(df,df1, direction = 'nearest')

> 언제 사용할까 ?
  프로세스에서 샘플링한 데이터로 작업하고 있으며 날짜 또는 시간이 정확히 일치 하지 않을 수 있을때 주로 사용한다.
