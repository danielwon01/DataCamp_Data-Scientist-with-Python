# Concatenate DataFrames

>데이터 테이블을 수직으로연결, 기본적으로 concat은 outer join을 default로 시행된다.  
 열이 다른 테이블을 연결 시 기본적으로 concat method는 결합하는 다른 테이블의 모든 열을 포함한다. 

> pd.concat([df,df1] , ignore_index = True , keys = [] )

* ignore_index = True는 일반적으로 concat은 index 값이 유지된 상태로 병합이 이루어지는데 ignore_index =True를 사용하면 index가 없이 병합이 이루어진다.  


* keys는 multi index 형태로 병합이 이루어진다. keys를 설정하게되면 ignore_index는 사용하지 못한다.   


```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.arange(16).reshape(4,4), index = ['A','B','C','D'] , columns = [1,7,3,5])

df1 = pd.DataFrame(np.arange(16).reshape(4,4), index = ['E','F','G','H'] , columns = [5,2,7,8])

display(df ,df1)
```

### Concat 실행

```python
df_c = pd.concat([df,df1])
df_c
```

### Concat(ignore_index =True)  - index 값 제거 

```python
df_ign = pd.concat([df,df1], ignore_index =True)
df_ign
```

### Concat(keys = [] )  - multi index 

```python
df_keys = pd.concat([df,df1], keys =[1 ,2])
df_keys
```

### Concat에서 join 설정 

```python
df_join  = pd.concat([df,df1], join = 'inner') # 같은 열인 7열만 병합되서 반환 
df_join
```

### Concat에서 join은 outer와 inner만 사용 가능하다.
