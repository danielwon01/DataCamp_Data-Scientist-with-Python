# Explicit indexes

```python
import numpy as np
import pandas as pd
```

```python
sample = pd.read_csv("Desktop/SampleSuperstore.csv")
sample.head(3)
```

#### set_index()  - index 지정

```python
sample.set_index("Country").head(3)
```

#### reset_index() - 인덱스를 재설정  
* drop =True 인수를 사용하면 인덱스를 삭제 할 수 있다.

```python
sample.reset_index(drop = True).head(3) 
```

#### 멀티인텍스, hierarchical indexes (여러 인덱스 설정)

```python
sample.set_index(["Country" , "State"]).head(3)
```
