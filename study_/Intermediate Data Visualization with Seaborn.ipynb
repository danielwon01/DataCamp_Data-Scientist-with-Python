# Intermediate Data Visualization with Seaborn
Seaborn은 matplotlib을 기반으로 통계 시각화 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action= "ignore")
```

```python
penguins = sns.load_dataset("penguins" )
penguins.head(3)
```

### 일변량 분석

```python
fig , ax  = plt.subplots()

penguins["bill_length_mm"].plot.hist()
```

```python
fig , ax  = plt.subplots()

sns.distplot(penguins["bill_length_mm"])
```

Seaborn API는 다른 인수를 사용하여 플롯 유형 사용자 정의를 지원 

```python
fig , ax  = plt.subplots()

sns.distplot(penguins["bill_depth_mm"],  hist = False , rug = True , kde_kws= {'shade' : True } )  # kde_kws 플롯을 음영처리하도록 kde함수에 지시
```

### Regression Plots in Seaborn( 상관분석 이변량분석)

```python
sns.regplot(x = "bill_length_mm", y= "bill_depth_mm", data = penguins)
```

```python
# lmplot 사용방법은 regplot과 유사하지만 단일변수를 변경하면서 여러그래프를 플로팅하는 패싯 등 여러 기능 사용가능 

sns.lmplot(x = "bill_length_mm", y= "bill_depth_mm", data = penguins, hue = "species", col = "sex")
```

### Using Seaborn Styles

```python
# seaborn 기본 테마를 설정하기 위해서는 sns.set() , pandas, matplotlib안에서도 연동

for style in ["white" , "dark", "whitegrid", "darkgrid", "ticks"] :
    sns.set_style(style)
    sns.distplot(penguins["body_mass_g"])
    plt.show()
```

```python
# despine()  축선 제거 
sns.distplot(penguins["body_mass_g"])
sns.despine(left = True)
```

### Colors in Seaborn

```python
#기본으로6개의 색상 제공

for p in sns.palettes.SEABORN_PALETTES :
    sns.set_palette(p)
    sns.displot(penguins["flipper_length_mm"])
    plt.show()
```

```python
for p in sns.palettes.SEABORN_PALETTES :
    sns.set_palette(p)
    sns.palplot(sns.color_palette())  # 팔레트 색상으로 보여주는 function
    plt.show()
```

```python
# circular colors 데이터가 범주형일때 유용
sns.palplot(sns.color_palette("Paired", 12))
```

```python
# sequential colors 데이터가 높은값에서 낮은값으로 일관된 경우에 유용
sns.palplot(sns.color_palette("Purples", 12))
plt.show()
```

```python
# diverging colors 높은값과 낮은 값이 모두 interesting할때 유용
sns.palplot(sns.color_palette("BrBG", 12 ))
```

### Customizing with matplotlib

```python
fig ,ax = plt.subplots()

sns.distplot(penguins["flipper_length_mm"], ax = ax) # seaborn안에 matplotlib ax를 객체를 넣어주어 커스터마이징 가능하다.
ax.set(xlabel = "flipper_length_mm", ylabel  = "Density", ylim = (0, 0.035)) 
```

```python
penguins.columns
```

```python
fig ,(ax0, ax1) = plt.subplots(2,1,figsize = (8,6) ,sharey =False) # sharey x,y축값을 동일시 할 것인지 

sns.regplot(x = "bill_length_mm", y= "bill_depth_mm", data = penguins, ax =ax0)
sns.regplot(x= "flipper_length_mm", y= "body_mass_g" , data = penguins, ax= ax1)
```

```python
median = penguins["flipper_length_mm"].median()
```

```python
median_b = penguins["body_mass_g"].median()
```

```python
fig ,(ax0, ax1) = plt.subplots(1,2,figsize = (8,6) ,sharey =False) 

sns.distplot(penguins["flipper_length_mm"] , ax = ax0)
sns.distplot(penguins["body_mass_g"], ax = ax1)

ax0.axvline(x = median, color = "r", label = "median", linestyle ="--", linewidth = 2)
ax1.axvline(x = median_b , color = "g" , label = "mean", linestyle = "-" , linewidth = 2)
```
