```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('data/Diabetes.csv')
df_original.head()
```

```python
# Pregnancies 와 outcome 컬럼을 제외한 컬럼들은 0의 값을 NaN으로 변경
cols = [c for c in df_original.columns if c not in ["Pregnancies", "Outcome"]]
df = df_original.copy()
df[cols] = df[cols].replace({0: np.NaN})
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Scatter Matrix - only works with numeric data

```python
pd.plotting.scatter_matrix(df , figsize = (12,10))
plt.show()
```

```python
df2 = df.dropna()
colors = df2['Outcome'].map(lambda x: "#44d9ff" if x else "#f95b4a")
pd.plotting.scatter_matrix(df2, figsize=(12,10), color=colors)
plt.show()
```

### Correlation Plots

```python
df.corr()
```

```python
sns.heatmap(df.corr())
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="viridis", fmt="0.2f")
plt.show()
```

### 2D Histograms

```python
df2 = pd.read_csv("data/height_weight.csv")
df2.info()
df2.describe()
```

```python
plt.hist2d(df2["height"], df2["weight"], bins=20, cmap="magma")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

### Contour plots

```python
hist, x_edge, y_edge = np.histogram2d(df2["height"], df2["weight"], bins=20)
x_center = 0.5 * (x_edge[1:] + x_edge[:-1])
y_center = 0.5 * (y_edge[1:] + y_edge[:-1])

plt.contour(x_center, y_center, hist, levels=4)
plt.xlabel("Height")
plt.ylabel("Weight");
```
