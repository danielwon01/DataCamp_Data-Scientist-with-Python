# Append 

>append는 DataFrame method 이기 때문에 테이블이 정렬한 상태에서 사용하여야 한다.

> merge, concat에 비교하면 append는 단순화된 연결이라고 할 수 있다.

> df.append(df2 )  - ignore_index는 지원 하지만 keys와 join은 지원하지 않는다. 항상 outer join형식으로 병합 

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.arange(12).reshape(4,3), index = ['A','B','C','D'] , columns = [1,7,3,])

df1 = pd.DataFrame(np.arange(12).reshape(4,3), index = ['E','F','G','H'] , columns = [5,2,7,])

display(df ,df1)
```

### Append 사용

```python
df_A = df.append(df1)
df_A
```
