# melt () 

> 넓은 형식에서 긴형식으로 테이블의 피벗을 해제합니다. 컴퓨터 친화적인 형식으로 변경하는 것

> df.melt(id_vars = [] ,value_vars ='', var_name= '' , value_name  = '')

>  id_vars 식별자 변수 또한 변경하지 않으려는 원래 데이터 세트의 열   
   value_vars =[] 원하는 데이터만 melt  
    melt하게 되면 variable과 value로 열 이름이 지정되는 데  var_name , value_name  = 열이름 지정

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("penguins" )
df.head(3)
```

### pivot_table 생성

```python
pivot = pd.pivot_table(df, index = ['sex','species'])
pivot
```

### melt 사용

```python
pivot.melt(id_vars='body_mass_g',var_name ='penguin', value_name = 'penguin_v' )
```
