# query () 

> 테이블에서 반환 할 행을 선택하는데 사용할 문자열을 받는다. SQL의 WHERE 문과 유사 

> df.query('col > = value')  
  df.query('col > value and(&) col < value')  
  df.query('col == " " or (col == " " and col > value') )  
  
> 여러가지 응용이 가능하다.

#### 라이브러리와 데이터 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("penguins" )
df.head(5)
```

### query 함수 사용

```python
M = df.query('sex == "Male"') # 문자열로 받아줘야함 
M.head(3)
```

```python
sp = df.query('species == "Gentoo" & sex == "Female"')
sp.head(3)
```

```python
ma = df.query('species == "Adelie" | (flipper_length_mm >= 210 & body_mass_g >= 455)')
ma.head(3)
```
