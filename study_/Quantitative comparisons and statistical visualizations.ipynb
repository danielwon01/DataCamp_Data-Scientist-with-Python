# Quantitative comparisons and statistical visualizations

#### 라이브러리 불러오기 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns

penguins = sns.load_dataset("penguins" )
penguins.head(3)
```

### Bar 차트  - 데이터에 대한 값의 크기를 막대로 보여주는 plot 
>막대 차트는 데이터의 분포를 표시하거나 데이터의 여러 하위 그룹에 걸쳐 데이터의 값을 비교하려는 경우에 사용됩니다. 

```python
fig , ax  = plt.subplots( figsize = (8,6) )

ax.bar(penguins["species"] , penguins["body_mass_g"] , width =0.4 )
```

### error-bar  - 에러바 (Errorbar, 오차막대)는 데이터의 편차를 표시하기 위한 그래프 형태
>  값의 표준편차와 같이 데이터 분포를 하나의 숫자로 요약

```python
# 바 차트에 에러바 추가 

fig , ax  = plt.subplots( figsize = (8,6) )

ax.bar(penguins["species"] , penguins["flipper_length_mm"] , width =0.4 , yerr = penguins["flipper_length_mm"].std() )

# yerr의 각 값들은 데이터 포인트의 위/아래 대칭인 오차로 표시됩니다.
```

### 히스토그램 
> 빈도 분포는 데이터 집합에서 서로 다른 각 값이 얼마나 자주 발생하는지 보여주는데, 히스토그램은 빈도 분포를 표시하는 데 가장 일반적으로 사용되는 그래프  


>연속형 측정값으로 값의 분포를 이해하고 이상치를 찾으려고 할때 사용합니다.

```python
fig , ax = plt.subplots(figsize = (8,6))

weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71,
          80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

weight1 = [70, 80, 64, 56, 77, 56, 77, 88, 31, 90, 45 ,
         65, 58, 79 ,  64 , 95 , 75 , 65 , 70 , 80 , 59 , 79]


ax.hist(weight , bins = 6 , histtype = "step" ,label = "A" )
ax.hist(weight1 , bins =6 , histtype = "step" ,label = "B" )

ax.set_xlabel("Weight")
ax.legend() # legend를 쓰기 위해서는 그래프를 만들때 label 지정 필요 
```

### boxplot  -데이터의 각 종류별로 사분위 수(quantile)를 표시하는 plot.
>  특정 데이터의 전체적인 분포를 확인하기 좋은 시각화 기법입니다.

> whisker - 상자의 좌우 또는 상화로 뻗어나가 선  
  박스 내부의 가로선 - 중앙값  
  lower whisker - 최솟값(중앙값 - 1.5 * IQR 보다 큰 데이터 중 가장 작은 값)  
  upper whisker - 최댓값(중앙값 + 1.5*IQR 보다 작은 데이터 중 가장 큰 값)  
 IQR(Inter Quartile Range) - 제3 사분위수 - 제1 사분위수   
 이상치 - lower whisker 보다 작은 데이터 또는 upper whisker 보다 큰 데이터

### scatter plots -두 개 변수 간의 관계를 나타내는 그래프 방법
